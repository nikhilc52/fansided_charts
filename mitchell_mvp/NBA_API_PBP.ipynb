```python
from nba_api.stats.endpoints import teamplayeronoffsummary
import pandas as pd
import time
from datetime import datetime, timedelta

start_date = datetime(2023, 12, 16)
end_date = datetime(2024, 2, 14)

onoff = teamplayeronoffsummary.TeamPlayerOnOffSummary(
    team_id = 1610612739, 
    date_from_nullable = start_date.strftime("%m/%d/%Y"), 
    date_to_nullable = start_date.strftime("%m/%d/%Y"))

onoff.get_data_frames()[1]

start_date += timedelta(days=1)

while start_date <= end_date:
    time.sleep(0.600)
    onoff = teamplayeronoffsummary.TeamPlayerOnOffSummary(
        team_id = 1610612739, 
        date_from_nullable = start_date.strftime("%m/%d/%Y"), 
        date_to_nullable = start_date.strftime("%m/%d/%Y"))   
    
    print("Game on " + start_date.strftime("%m/%d/%Y"))
    
    start_date += timedelta(days=1)
    combined = combined.append(onoff.get_data_frames()[1])
    
combined.to_csv('CavsOnly20On.csv', index=False)
```
